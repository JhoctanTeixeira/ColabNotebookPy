```
lista1 = [1, 2, 5, 8, 11, 13,17, 14]
```

4) Crie uma lista chamada lista3 contendo a idade de 5 pessoas. Após, percorra a lista
utilizando o for e verifique quantas delas atingiram a maioridade e quantas ainda não
atingiram. Guarde as informações nas variáveis maioridade e menoridade.


```
lista3 = [17,20,22,32,44]
maioridade = 0
menoridade = 0
```

```
for w in lista3:
 if w >= 18:
   maioridade = maioridade + 1
 else:
   menoridade += 1 
```

```
print('{} pessoa(s) tem mais de 18 anos. \n{} pessoa(s) tem menos de 18 anos'.format(maioridade, menoridade))
```
