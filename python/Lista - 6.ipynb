######3. Peça ao usuário para digitar uma frase. Escreva um programa para verificar a quantidade de letras maiúsculas e minúsculas da frase digitada pelo usuário. Utilize dicionários para salvar a quantidade de cada uma.
######Dicionario = {‘maiusculas’: 0, ‘minusculas’: 0, ‘outros’: 0}


```
dicionário = {}
minuscula = 0
maiuscula= 0
outros = 0

frase = (input('Digite uma frase: '))
for letra in frase:
     if letra.islower():
        minuscula = minuscula + 1
     elif letra.isupper():
        maiuscula = maiuscula + 1
     else:
          outros = outros + 1
dicionário['maiuscula'] = maiuscula
dicionário['minuscula'] = minuscula
dicionário
  
```
