11.	Faça um programa que pergunte ao usuário seu sexo, altura e peso e retorne o seu índice de massa corporal (IMC) através da seguinte fórmula: 
    - **imc =  peso / (altura * altura)**

Após o calculo do IMC imprima uma mensagem informando o IMC juntamente com as informações de sexo, altura e peso.


```python
# Função input para string
sexo = input('Qual seu sexo: ')
```

```python
# Função input para valores reais (float)
altura = float(input('Qual sua altura:'))
```

```python
# Função input para valores inteiros (int)
peso = int(input('Qual seu peso: '))
```

```python
# É uma boa prática em programação que variáveis nunca iniciem com letras
# maiúsculas.
imc = peso / (altura ** 2)
print('O imc dessa pessoa é de {:.1f}'.format(imc))
```

```python
print('O peso é', peso)
print('O sexo é', sexo)
print('A altura é', altura)
print('o imc é', imc)
```
