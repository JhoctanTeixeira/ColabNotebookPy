#####4. Escreva um programa que apresente três opções de ações numa conta do banco, que serão apresentadas a seguir. O saldo deve iniciar em R$ 0,00. A cada saque ou depósito o valor do saldo deve ser atualizado. A condição de saída do programa é se o usuário digitar “sair”.

```
Saldo = 0
Saque = 0
Depósito = 0
resposta = '' 
while resposta != 'Sair':
  resposta = input("Escolha uma dás 3 opções: \n \n A) Saldo; \n B) Saque; \n C) Depósito. \n \nDigite Sair para pode sair das opções. \n \n")
  if resposta == 'A':
    print('\n Saldo atual: \n BRL R$ {},00 \n'.format(Saldo))
  if resposta == 'B':
    Saque = int(input('Qual o valor do saque? '))
    if Saque < Saldo or Saque == Saldo:
      Saldo -= Saque
      print('\n Saldo atual: \nBRL R$ {},00 \n'.format(Saldo))
    else:
      print('\n OPERAÇÃO DE INVÁLIDA: \n Saldo inferior ao valor escolhido para saque. \n ')
  if resposta == 'C':
    Depósito = int(input('Qual o valor do depósito? \n'))
    Saldo += Depósito
    print('\n Saldo atual: \nBRL R$ {},00 \n'.format(Saldo))
```
