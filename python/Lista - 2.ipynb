6. Crie um programa que mostre a tabuada de um número que o usuário escolher, utilizando
o laço for.


```
num = int(input('Digite um número para ver sua tabuada: '))
for c in range(1, 11):
  print('{} x {:2} = {}'.format(num, c, num*c))
```
