4. Escreva uma função que recebe uma lista de palavras (strings) e retorna o número de
palavras na lista que tem comprimento 5.

```
lista = ['carros','peixe','galinha','cachorro','bode','avião','tornado']
def letras(lista):
  contador = 0
  for t in lista:
    if len(t) >= 5:
      contador += 1
  return (contador)
result = letras(lista)
print(result)
```
